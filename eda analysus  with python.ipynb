```python
correct_mapping = {
    # Clothing
    "T-shirt": "Clothing","Shirt":"Clothing","Shorts":"Clothing",
    "Hoodie":"Clothing","Pants":"Clothing","Socks":"Clothing",
    "Jeans":"Clothing","Blouse":"Clothing","Skirt":"Clothing",
    "Sweater":"Clothing","Dress":"Clothing",
    
    # Accessories
    "Sunglasses":"Accessories","Gloves": "Accessories","Jewelry":"Accessories",
    "Hat":"Accessories", "Handbag":"Accessories","Backpack":"Accessories",
    "Belt":"Accessories","Bag": "Accessories","Scarf":"Accessories",
    
    # Electronics
    "Laptop":"Electronics","Watch":"Electronics","Headphones":"Electronics","Phone":"Electronics",
    
    # Footwear
    "Shoes":"Footwear","Sandals":"Footwear","Sneakers":"Footwear","Boots":"Footwear",
    
    # Outerwear
    "Coat":"Outerwear","Jacket":"Outerwear"}
```

```python
correct_mapping
```

```python
df.head()
```

```python
df["Category"]=df["Item Purchased"].map(correct_mapping)
```

```python
df['Category'].unique()
```

```python
df.isnull().sum()
```

```python
df.groupby("Category")["Size"].unique()
```

```python
#treating nullvalue or replace all values in the electonics category
df.loc[df['Category']=='Electronics','Size']="notapplicable"
```

```python
#treating nullvalue or replace all values in the footwere category
df.loc[df['Category']=='Footwear','Size']="not avalable"
```

```python
#treating nullvalue or replace all values in the Accessories  category
df.loc[df['Category']=='Accessories','Size']="al free"
```

```python
#treating nullvalue or replace all values in the clothing  category
df.loc[(df["Category"]=="Clothing") & (df["Size"].isnull()),"Size"]=df[df["Category"]=="Clothing"]["Size"].mode()[0]
```

```python
#fill null values nullvalue or replace all values in the review ratings
df["Review Rating"]=df["Review Rating"].fillna(df.groupby('Item Purchased')["Review Rating"].transform('mean'))
```

```python
#filling Purchase Amount with (0)
df['Previous Purchases']=df['Previous Purchases'].fillna(0)
```

```python
#fill null values nullvalue or replace all values in the 
df["Purchase Amount (USD)"]=df["Purchase Amount (USD)"].fillna(df.groupby('Item Purchased')["Purchase Amount (USD)"].transform('mean'))
```

```python
import pyodbc
from urllib.parse import quote_plus
from sqlalchemy import create_engine

driver = "ODBC Driver 17 for SQL Server"

connection_string = quote_plus(
    f"DRIVER={{{driver}}};"
    f"SERVER=localhost;"
    f"DATABASE=dataanalyst_analyst_project;"
    f"Trusted_Connection=yes;"
)

# create engine
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={connection_string}")

print("successfully executed")
```

```python
table_name="costumer_behaviour_analysis"
df.to_sql(table_name, engine, if_exists="replace", index=False)
```
